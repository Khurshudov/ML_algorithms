```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = []
y = []
for i in range(50):
    X.append([rd.random() * 4, rd.random() * 10])
    y.append(-1)
    
for i in range(50):
    X.append([6 + rd.random() * 4, rd.random() * 10])
    y.append(1)
    
X = np.array(X)
y = np.array(y)
```

```python
def L(w, x, y):
    return 1 + np.exp(-np.dot(w, x) * y)

def Ld(w, x, y):
    if np.dot(w, x) * y < 0:
        return 1
    return 0

def find_x(w, y):
    return (w[2] - w[1] * y) / w[0]

def train(X, y, eta=0.02, lambd = 0.9):
    X = np.append(X, np.array([-1] * len(X)).reshape(len(X), 1), axis=1)
    w = np.array([0.1] * len(X[0]))
    Q = 1 / len(X) * sum([ L(w, X[i], y[i]) for i in range(len(X)) ])
    
    kf = 1
    
    eta = 1
    
    n_iter = 0
    while (n_iter < 10000) and kf > 1e-5:
        
        eta = max(eta - 1 / (n_iter + 1), 0.03)
        
        Q_prev = Q
        n_iter += 1
        i = rd.randint(0,99)
        x = X[i]
        eps = L(w, x, y[i])
        w = w + eta * Ld(w, x, y[i]) * x * y[i]
        Q = (1-lambd) * Q + lambd * eps
        kf = (abs(Q_prev - Q) / abs(max(Q_prev, Q)))
    print('n_iter:', n_iter)
    return w
```

```python
w = train(X, y)

plt.scatter(X[:,0], X[:,1], c=y)
plt.plot([find_x(w, 0), find_x(w, 10)], [0, 10])
```
