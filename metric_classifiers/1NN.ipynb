```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = load_iris()
X = data.data[:,2:]
y = data.target
```

```python
def euc(x, y):
    x = np.array(x)
    y = np.array(y)
    return sum((x-y) ** 2)

def nearest_index(x, X):
    min_euc = euc(x, X[0])
    min_idx = 0
    for idx, elem in enumerate(X):
        if min_euc > euc(x, elem):
            min_euc = euc(x, elem)
            min_idx = idx
    return min_idx

def predict(x, X, y):
    return y[nearest_index(x,X)]

def color(col):
    if str(col) == '0':
        return 'red'
    elif str(col) == '1':
        return 'green'
    return 'blue'
```

```python
%%time

plt.figure(1, figsize=(8, 6))
plt.scatter(X[:,0], X[:,1], c=list(map(color, y)))

step = 0.1

for i in np.arange(1, 7, step=step):
    for j in np.arange(0, 2.5, step=step):
        col = str(predict([i,j], X, y))

        plt.plot(i, j, c=color(col), marker='o', fillstyle='none')
```
